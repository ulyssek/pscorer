##### Welcome to P-scorer, with this class, you can obtain, from a pdf file the list of all (Inch'Allah) p_values stored in it and plot there distribution.

#### Let's see how it works.


# Import Section

```python

from pscorer import PScorer # The class that will be used to explore p_values

```

# Plotting the distribution of two articles

```python

files_path = './' ## Put here the path to your files. '.' if the directory is the same

s = PScorer(files_path=files_path) # Creating the scorer, then give it article to explore

s.distrib('zns.pdf') # Will plot the distribution for the article zns.pdf
s.distrib("churchland.pdf",title='van Vugt 2018') # Will plot the distribution for the article van_vugt.pdf

print("was Easy")
```

# Plotting the combined distribution of two articles

```python
s = PScorer()

s.distrib(['zns.pdf',"churchland.pdf"])
print("Much easier")
```

# Getting the full p_value list

```python
# Get the p_value list for your article
# Note that your p_values are extracted and formated (every article uses a different p_value format)

s = PScorer()

p_values = s.get_p_values('zns.pdf')
print(p_values)

## Important note : 
## when a p_value is found to be <.05 or exactly .05, a random value is used between .01 and .05
## when a p_value is found to be <.01 or exactly .01, a random value is used between .001 and .01
## when a p_value is found to be <.001 or exactly .001, a random value is used between 0. and .001
## Therefor, the value given by get_p_values and the graph obtain with distrib can change from one run to another
```

# Getting the raw p_value list

```python
# The p_values are displayed as they are writen in the article, without any later formating

s = PScorer()

p_values = s._get_raw_p_values('zns.pdf')
print(p_values)
```

# Getting the raw content of the pdf file

```python
# The parameters v_talky changes the behaviour of the function to write the content of the pdf


s = PScorer()

s.get_p_values('zns.pdf',v_talky=True)
print("Not cool")
```

# Custom article

##### Not all article use the same format, unfortunatly, for some of them, a custom regexp and analysis is required

```python

s = PScorer()

s.distrib('powers2017.pdf',title="Ugly distribution, though the article is very good")

print("The distribution seems ugly since most of the p-values are stored in the format 5.45x10-7\n")
print("""Two things need to be done here. First, a custom search pattern must be used to include this values.
Then, a custom converting function must be used to extract them""")

custom_pattern = '[P|p][ ]*[=|<]*[ ]*[0-9]?[.][0-9]+[×10\n[0-9]+]?' # Pattern fitted for this article

def custom_convert(self,d):
    e = d.split('×')
    if len(e) == 1:
        convert_d = self._convert(d)
        return float(convert_d)
    else:
        a = e[0]
        b = e[1].split('\n')[-1]
        return float(a)*10**(-float(b))

s.distrib('powers2017.pdf',search_pattern=custom_pattern,title="Corrected distribution including all the p-values",convert_function=custom_convert)

print("Don't see any generic to catch these missing values :'(")
```
